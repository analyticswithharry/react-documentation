## 3. Display Output with JSX

JSX is how React shows output on the screen.

You can use:

- plain text
- HTML-like tags
- JavaScript expressions inside `{ }`

This is the main idea of output in React: **return JSX from a component**.

```python
function App() {
  // Store a normal JavaScript variable
  const name = 'Hemant';

  // Store a number
  const score = 95;

  // Return JSX to show output on screen
  return (
    <div>
      {/* Plain text output */}
      <h1>Welcome to React</h1>

      {/* Show a variable with curly braces */}
      <p>My name is {name}</p>

      {/* Show a number with curly braces */}
      <p>My score is {score}</p>

      {/* Show an expression */}
      <p>Next score will be {score + 5}</p>
    </div>
  );
}

export default App;
```

## 4. Handle User Input with Events

React listens to browser events like:

- button click
- input change
- form submit

When something happens, React calls a function that you wrote.

```python
function App() {
  // Function that runs when the button is clicked
  function handleClick() {
    alert('Button clicked');
  }

  // Function that runs when input value changes
  function handleChange(event) {
    // event.target.value is the current text inside the input
    console.log('You typed:', event.target.value);
  }

  return (
    <div>
      {/* Click event */}
      <button onClick={handleClick}>Click me</button>

      {/* Change event */}
      <input
        type="text"
        placeholder="Type here"
        onChange={handleChange}
      />
    </div>
  );
}

export default App;
```

## 5. Manage Input with State

Now we connect input to state.

This is called a **controlled input**.

That means:

- React stores the value in state
- the input shows that state
- when the user types, state updates
- output changes immediately

```python
import { useState } from 'react';

function App() {
  // Create state to store the current input text
  const [message, setMessage] = useState('');

  function handleChange(event) {
    // Update state when user types
    setMessage(event.target.value);
  }

  return (
    <div>
      <h1>Input and Output Example</h1>

      {/* Input reads value from state */}
      <input
        type="text"
        value={message}
        onChange={handleChange}
        placeholder="Type your message"
      />

      {/* Output shows the same state */}
      <p>You typed: {message}</p>
    </div>
  );
}

export default App;
```

## 6. Build a Controlled Form

A controlled form means React keeps the form values in state.

We will use:

- text input
- textarea
- select
- submit button
- `event.preventDefault()` to stop page reload

```python
import { useState } from 'react';

function App() {
  // Store all form fields in state
  const [formData, setFormData] = useState({
    name: '',
    bio: '',
    role: 'student',
  });

  // Store submitted result separately
  const [submittedData, setSubmittedData] = useState(null);

  function handleChange(event) {
    // Get input name and value from the changed field
    const { name, value } = event.target;

    // Update only the changed field
    setFormData((currentData) => ({
      ...currentData,
      [name]: value,
    }));
  }

  function handleSubmit(event) {
    // Stop normal browser form submit behavior
    event.preventDefault();

    // Save submitted data
    setSubmittedData(formData);
  }

  return (
    <div>
      <form onSubmit={handleSubmit}>
        <input
          name="name"
          value={formData.name}
          onChange={handleChange}
          placeholder="Enter your name"
        />

        <textarea
          name="bio"
          value={formData.bio}
          onChange={handleChange}
          placeholder="Write a short bio"
        />

        <select
          name="role"
          value={formData.role}
          onChange={handleChange}
        >
          <option value="student">Student</option>
          <option value="developer">Developer</option>
          <option value="designer">Designer</option>
        </select>

        <button type="submit">Submit form</button>
      </form>

      {submittedData && (
        <div>
          <h2>Submitted Output</h2>
          <p>Name: {submittedData.name}</p>
          <p>Bio: {submittedData.bio}</p>
          <p>Role: {submittedData.role}</p>
        </div>
      )}
    </div>
  );
}

export default App;
```

## 7. Render Lists and Conditional UI

React often shows arrays of data and sometimes shows UI only when a condition is true.

We use:

- `map()` for lists
- ternary or `&&` for conditionals

```python
function App() {
  // A simple array of menu items
  const menuItems = ['Home', 'About', 'Contact'];

  // A condition value
  const isLoggedIn = true;

  return (
    <div>
      <h1>Menu</h1>

      {/* Render list items with map */}
      <ul>
        {menuItems.map((item) => (
          <li key={item}>{item}</li>
        ))}
      </ul>

      {/* Conditional UI */}
      {isLoggedIn ? <p>Welcome back!</p> : <p>Please log in.</p>}
    </div>
  );
}

export default App;
```

## 8. Pass Data with Props

Props are how a parent component sends data to a child component.

Think of props like function arguments for UI components.

```python
// Child component receives props from parent
function GreetingCard({ name, role }) {
  return (
    <div>
      <h2>{name}</h2>
      <p>{role}</p>
    </div>
  );
}

function App() {
  // Parent sends values into child component
  return (
    <div>
      <GreetingCard name="Hemant" role="React learner" />
      <GreetingCard name="Asha" role="Frontend student" />
    </div>
  );
}

export default App;
```

## 9. Add React Router DOM

Now we learn `react-router-dom`.

These are the most important router pieces:

- `BrowserRouter` → turns on routing in the app
- `Routes` → holds all route rules
- `Route` → one route rule
- `Link` → moves between pages without full reload

```python
// main.jsx
import ReactDOM from 'react-dom/client';

// Import BrowserRouter from react-router-dom
import { BrowserRouter } from 'react-router-dom';

import App from './App';

ReactDOM.createRoot(document.getElementById('root')).render(
  // Wrap the whole app with BrowserRouter
  <BrowserRouter>
    <App />
  </BrowserRouter>
);

// App.jsx
import { Routes, Route, Link } from 'react-router-dom';

function Home() {
  return <h2>Home Page</h2>;
}

function About() {
  return <h2>About Page</h2>;
}

function App() {
  return (
    <div>
      {/* Link lets user move to another route */}
      <nav>
        <Link to="/">Home</Link>
        <Link to="/about">About</Link>
      </nav>

      {/* Routes contains Route items */}
      <Routes>
        {/* When URL is / show Home */}
        <Route path="/" element={<Home />} />

        {/* When URL is /about show About */}
        <Route path="/about" element={<About />} />
      </Routes>
    </div>
  );
}

export default App;
```

## 10. Create Routes and Navigation

A small app usually has multiple pages like:

- Home
- About
- Contact

Navigation is just a group of `Link` components.

```python
import { Routes, Route, Link } from 'react-router-dom';

function Home() {
  return <h2>Home Page</h2>;
}

function About() {
  return <h2>About Page</h2>;
}

function Contact() {
  return <h2>Contact Page</h2>;
}

function App() {
  return (
    <div>
      <nav>
        {/* Navigation links */}
        <Link to="/">Home</Link> |{' '}
        <Link to="/about">About</Link> |{' '}
        <Link to="/contact">Contact</Link>
      </nav>

      <Routes>
        <Route path="/" element={<Home />} />
        <Route path="/about" element={<About />} />
        <Route path="/contact" element={<Contact />} />
      </Routes>
    </div>
  );
}

export default App;
```

## 11. Read Route Parameters

Sometimes a route needs dynamic values.

Example:

- `/user/1`
- `/user/25`

We use `useParams()` to read that value from the URL.

```python
import { Routes, Route, Link, useParams } from 'react-router-dom';

function Home() {
  return (
    <div>
      <h2>Home Page</h2>
      <Link to="/user/101">Open User 101</Link>
    </div>
  );
}

function UserPage() {
  // Read the id value from the URL
  const { id } = useParams();

  return <h2>User ID from URL: {id}</h2>;
}

function App() {
  return (
    <Routes>
      <Route path="/" element={<Home />} />
      <Route path="/user/:id" element={<UserPage />} />
    </Routes>
  );
}

export default App;
```

## 12. Handle Not Found Routes

When the user opens a page that does not exist, we can show a custom 404 page.

We do that with a wildcard route using `*`.

```python
import { Routes, Route } from 'react-router-dom';

function Home() {
  return <h2>Home Page</h2>;
}

function NotFound() {
  return <h2>404 - Page Not Found</h2>;
}

function App() {
  return (
    <Routes>
      <Route path="/" element={<Home />} />

      {/* Wildcard route catches unknown URLs */}
      <Route path="*" element={<NotFound />} />
    </Routes>
  );
}

export default App;
```

## 13. Organize Components for a Small App

When your app grows, split code into folders.

A small beginner structure could look like this:

- `src/components/`
- `src/pages/`
- `src/App.jsx`
- `src/main.jsx`

Below is one simple example.

```python
// src/main.jsx
// Start the app and enable router
import ReactDOM from 'react-dom/client';
import { BrowserRouter } from 'react-router-dom';
import App from './App';

ReactDOM.createRoot(document.getElementById('root')).render(
  <BrowserRouter>
    <App />
  </BrowserRouter>
);

// src/App.jsx
// Main app file with navigation and routes
import { Routes, Route, Link } from 'react-router-dom';
import HomePage from './pages/HomePage';
import AboutPage from './pages/AboutPage';
import ContactPage from './pages/ContactPage';

function App() {
  return (
    <div>
      <nav>
        <Link to="/">Home</Link> |{' '}
        <Link to="/about">About</Link> |{' '}
        <Link to="/contact">Contact</Link>
      </nav>

      <Routes>
        <Route path="/" element={<HomePage />} />
        <Route path="/about" element={<AboutPage />} />
        <Route path="/contact" element={<ContactPage />} />
      </Routes>
    </div>
  );
}

export default App;

// src/pages/HomePage.jsx
// A page component can hold inputs, output, forms, and child components
function HomePage() {
  return <h2>Home Page</h2>;
}

export default HomePage;
```
